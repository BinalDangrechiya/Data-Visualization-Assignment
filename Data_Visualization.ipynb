Q 4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical variable across different categories.

Generating a Dataset with Categories and Numerical Values

Scenario: Let's create a dataset simulating the average monthly income of individuals based on their educational level.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
education_levels = ['High School', 'Bachelors', 'Masters', 'PhD']
num_samples = 100

data = {
    'Education': np.random.choice(education_levels, num_samples),
    'Income': np.random.normal(loc=60000, scale=20000, size=num_samples)
}
df = pd.DataFrame(data)

print(df.head())
```

Visualizing the Distribution

Box Plot: A box plot is a great way to visualize the distribution of a numerical variable across different categories.

```python
plt.figure(figsize=(10, 6))
plt.boxplot([df['Income'][df['Education'] == level] for level in education_levels])
plt.xticks(range(1, len(education_levels) + 1), education_levels)
plt.xlabel('Education Level')
plt.ylabel('Income')
plt.title('Income Distribution by Education Level')
plt.show()
```

Q 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

I'd be glad to help you generate a synthetic dataset with correlated features and visualize its correlation matrix using a heatmap in Python with Seaborn.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(42)
num_samples = 1000
num_features = 5

X = np.random.randn(num_samples, num_features)

correlation_matrix = np.array([[1.0, 0.8, 0.6, 0.4, 0.2],
                              [0.8, 1.0, 0.7, 0.5, 0.3],
                              [0.6, 0.7, 1.0, 0.6, 0.4],
                              [0.4, 0.5, 0.6, 1.0, 0.5],
                              [0.2, 0.3, 0.4, 0.5, 1.0]])

L = np.linalg.cholesky(correlation_matrix)
correlated_data = X @ L.T

# Create a Pandas DataFrame from the correlated data
df = pd.DataFrame(correlated_data, columns=['Feature 1', 'Feature 2', 'Feature 3', 'Feature 4', 'Feature 5'])

# Calculate the correlation matrix of the DataFrame
corr_matrix = df.corr()

# Visualize the correlation matrix using a heatmap
sns.set_theme()  # Set a theme for the plot
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

A heatmap is a graphical representation of data where values are represented by colors. It's often used to visualize patterns or trends in data.

```python
np.random.seed(20)
data_sales = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}

df_sales = pd.DataFrame(data_sales)

# Pivot the data for heatmap
sales_pivot = df_sales.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='mean')

# Create heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(sales_pivot, cmap='coolwarm', annot=True, fmt=".0f", linewidths=0.5)
plt.title("Sales Variation Across Different Months and Days")
plt.show()
```

```python
np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

plt.scatter(df['GDP'], df['Population'], s=df['Population'] * 10, alpha=0.5)
plt.title('Population vs. GDP')
plt.xlabel('GDP')
plt.ylabel('Population')
plt.grid(True)

for i, row in df.iterrows():
    plt.annotate(row['Country'], (row['GDP'], row['Population']))

plt.show()
```

Q 1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
import numpy as np
from bokeh.plotting import figure, output_file, show

x = np.linspace(0, 10, 1000)
y = np.sin(x)

# Create the plot
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')

# Add a line to the plot
p.line(x, y, line_width=2)

# Output the plot
output_file("sine_wave.html")
show(p)
```

Q 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

```python
import pandas as pd
from bokeh.models import ColumnDataSource

# Generate random data
np.random.seed(42)
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
sizes = np.random.randint(10, 50, N)
colors = ["#%06x" % np.random.randint(0, 0xFFFFFF) for _ in range(N)]

# Create a DataFrame for easier data handling
data = {'x': x, 'y': y, 'sizes': sizes, 'colors': colors}
df = pd.DataFrame(data)

# Create a ColumnDataSource
source = ColumnDataSource(df)

# Create the plot
p = figure(title="Scatter Plot with Varying Sizes and Colors")

# Add the scatter plot glyph
p.circle('x', 'y', source=source, size='sizes', color='colors', alpha=0.5)

# Output the plot
output_file("scatter_plot.html")
show(p)
```

Q 4. Create a Bokeh histogram to visualize the distribution of the given data.

```python
import numpy as np
from bokeh.plotting import figure, output_file, show

# Generate random data
data_hist = np.random.randn(1000)

# Calculate histogram
hist, edges = np.histogram(data_hist, bins=30)

# Create the figure
p = figure(title="Histogram of Random Data", x_axis_label='Value', y_axis_label='Frequency')

# Add histogram bars
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color='navy', line_color='black')

# Output the plot
output_file("histogram.html")
show(p)
```

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.models import LinearColorMapper, ColorBar, BasicTicker
from bokeh.io import output_notebook
from bokeh.palettes import Viridis256

# Display inline in Jupyter notebooks (optional)
output_notebook()

# Generate random data
data_heatmap = np.random.rand(10, 10)

# Create the color mapper
color_mapper = LinearColorMapper(palette=Viridis256, low=data_heatmap.min(), high=data_heatmap.max())

# Create the figure
p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y')

# Add the heatmap
p.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, color_mapper=color_mapper)

# Add a color bar
color_bar = ColorBar(color_mapper=color_mapper, ticker=BasicTicker(), location=(0, 0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```
